```python
df1=pd.read_csv("all_data.csv")
df1.head(400)
```

```python
#augment with additional columns
```

```python
#get rid of nans
```

```python
nan_df= df1[df1.isna().any(axis=1)]
nan_df.head()
df1=df1.dropna(how='all')
```

```python
##find or and delete
df1=df1[df1['Order Date'].str[0:2]!='Or']
```

```python
###add month column
```

```python
df1["Month"]=df1['Order Date'].str[0:2]
df1['Month']=df1['Month'].astype('int32')
df1.head(1000)
```

```python
#add a sales column
#correct to correct type
```

```python
df1["Quantity Ordered"]=pd.to_numeric(df1['Quantity Ordered'])
df1["Price Each"]=pd.to_numeric(df1["Price Each"])
```

```python
df1["Sales"]=df1['Quantity Ordered'] * df1['Price Each']

df1.head()
```

```python
results=df1.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt
```

```python
months=range(1,13)
plt.bar(months,results["Sales"])
plt.xticks(months)
plt.ylabel("sales in $")
plt.xlabel("Month number")
plt.show()
```

```python
##what city had the most ammount ofn sales
```

```python
#first must add city column
```

```python
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]
df1['City']=df1["Purchase Address"].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
df1.head()
```

```python
result=df1.groupby('City').sum()
result
```

```python
import matplotlib.pyplot as plt
cities= [city for city, df in df1.groupby('City')]
results=df1.groupby('City').sum()
plt.bar(cities,results["Sales"])
plt.xticks(cities, rotation='vertical',size=8)
plt.ylabel("sales in $")
plt.xlabel("City")
plt.show()
```

```python
df1["Order Date"]=pd.to_datetime(df1['Order Date'])
```

```python
##what products are most often sold together
```

```python
df2=df1[df1['Order ID'].duplicated(keep=False)]
```

```python
df2.head(20)
```

```python
df2['Grouped']=df2.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
```

```python
df2.head()
```

## 

```python
df2=df2[['Order ID','Grouped']].drop_duplicates()
```

```python
df2.head()
```

```python
from itertools import combinations
from collections import Counter
```

```python
count=Counter()
for row in df2['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
for key,value in count.most_common(10):
    print(key,value)
```
